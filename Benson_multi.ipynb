```python
%%manim -p -ql SetupTest

class SetupTest(Scene):
    def construct(self):
        # Green's equation
        # Use MathTex (or Tex) in recent manim versions — TextMobject was removed/deprecated.
        greens_eq = MathTex(
            r"\oint_C \mathbf{F} \cdot d\mathbf{r} = \iint_D \left( \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} \right) dA",
            tex_to_color_map={"C": BLUE, "D": GREEN}
        )
        greens_eq.scale(1)
        self.play(Write(greens_eq))
        self.wait(2)
        self.play(FadeOut(greens_eq))

```

Creating the scene (vector field and line integral)



```python
%%manim -p -ql Scene1

# 1) Scene 1 — Establish the world: vector field + curve \(C\)
class Scene1(Scene):
    def construct(self):
        # Create a NumberPlane for the grid
        plane = NumberPlane(x_range=[-5, 5, 1], y_range=[-3, 3, 1], faded_line_ratio=0)
        self.add(plane)

        # Define the vector function
        def vector_func(point):
            x, y, z = point
            # Example: Rotational field (-y, x)
            return np.array([-y, x, 0])

        # Create a VGroup to hold all the vector arrows
        vectors = VGroup()
        # Iterate over the grid to create individual arrows
        for x in np.arange(plane.x_range[0], plane.x_range[1] + plane.x_range[2], plane.x_range[2]):
            for y in np.arange(plane.y_range[0], plane.y_range[1] + plane.y_range[2], plane.y_range[2]):
                start_point = plane.coords_to_point(x, y)
                vec_value = vector_func(np.array([x, y, 0]))

                # Scale the vectors for better visibility (adjust 0.3 as needed)
                magnitude = np.linalg.norm(vec_value)
                if magnitude > 0:
                    scaled_vec = vec_value / magnitude * 0.3  # Normalize and scale
                else:
                    scaled_vec = np.array([0, 0, 0])

                end_point = start_point + scaled_vec
                arrow = Arrow(start_point, end_point, buff=0, color=BLUE, max_tip_length_to_length_ratio=0.25, stroke_width=2)
                vectors.add(arrow)

        # Animate the vector field and the curve
        self.play(Create(vectors), run_time=3)

        # Create a closed curve C (circle)
        circle = Circle(radius=2, color=YELLOW)
        self.play(Create(circle), run_time=2)
        self.wait(2)
```
